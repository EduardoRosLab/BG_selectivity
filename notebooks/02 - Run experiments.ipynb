```python
%matplotlib inline
from matplotlib import pyplot as plt
import nest
import numpy as np
import itertools
```

```python
from thesis_package.experiment import ExperimentSSTS
```

```python
e = ExperimentSSTS()
times = 40 #1 #100
e.run(times=times)
```

```python
import pickle
import time

save = {
    'times': times,
    'da_values': e.da_values,
    'hd_values': e.hd_values,
    
    'results_ss_hd_msn': e.results_ss_hd_msn,
    'results_ts_hd_msn': e.results_ts_hd_msn,
    'results_ss_hd_snr': e.results_ss_hd_snr,
    'results_ts_hd_snr': e.results_ts_hd_snr,

    'signal_hd_msn': e.signal_hd_msn,
    'signal_hd_snr': e.signal_hd_snr,
    
    'time_range': e.time_range,
}

folder = 'saved_experiments'
file_name = 'experiment_dump'
time = time.strftime('%Y-%m-%dT%H:%M:%S', time.gmtime())
with open('{}/{}_{}.pickle'.format(folder, file_name, time), 'wb') as f:
    pickle.dump(save, f)
with open('{}/{}.pickle'.format(folder, file_name), 'wb') as f:
    pickle.dump(save, f)
```

```python
# Plotting HD

plt.rcParams['figure.figsize'] = [10.0, 5.0]

fig, axs = plt.subplots(1,2)

axs[0].set_title('Transient selection', fontsize=16)
axs[0].set_xticks(range(len(e.hd_values)))
axs[0].set_xticklabels(['{:.1f}'.format(v) if i%2==0 else '' for i, v in enumerate(e.hd_values)])
axs[0].set_xlabel('\nHuntington disease level')
axs[0].set_yticks(range(len(e.da_values)))
axs[0].set_yticklabels(['{:.1f}'.format(v) if i%2==0 else '' for i, v in enumerate(e.da_values)])
axs[0].set_ylabel('Dopamine level')
cax = axs[0].contourf(e.results_ts_hd_msn, origin='lower', extent=[0,1,0,1])
fig.colorbar(cax, ax=axs[0], fraction=0.0485, pad=0.04)

axs[1].set_title('Steady-state selection', fontsize=16)
axs[1].set_xticklabels(['{:.1f}'.format(v) for v in e.hd_values])
axs[1].set_xlabel('\nHuntington disease level')
axs[1].set_yticklabels(['{:.1f}'.format(v) if i%2==0 else '' for i, v in enumerate(e.da_values)])
axs[1].set_ylabel('Dopamine level')
cax = axs[1].contourf(e.results_ss_hd_msn, origin='lower', extent=[0,1,0,1])
fig.colorbar(cax, ax=axs[1], fraction=0.0485, pad=0.04)

fig.tight_layout()



# Plotting HD

plt.rcParams['figure.figsize'] = [10.0, 5.0]

fig, axs = plt.subplots(1,2)

axs[0].set_title('Transient selection', fontsize=16)
axs[0].set_xticks(range(len(e.hd_values)))
axs[0].set_xticklabels(['{:.1f}'.format(v) if i%2==0 else '' for i, v in enumerate(e.hd_values)])
axs[0].set_xlabel('\nHuntington disease level')
axs[0].set_yticks(range(len(e.da_values)))
axs[0].set_yticklabels(['{:.1f}'.format(v) if i%2==0 else '' for i, v in enumerate(e.da_values)])
axs[0].set_ylabel('Dopamine level')
cax = axs[0].contourf(e.results_ts_hd_snr, origin='lower', extent=[0,1,0,1])
fig.colorbar(cax, ax=axs[0], fraction=0.0485, pad=0.04)

axs[1].set_title('Steady-state selection', fontsize=16)
axs[1].set_xticklabels(['{:.1f}'.format(v) for v in e.hd_values])
axs[1].set_xlabel('\nHuntington disease level')
axs[1].set_yticklabels(['{:.1f}'.format(v) if i%2==0 else '' for i, v in enumerate(e.da_values)])
axs[1].set_ylabel('Dopamine level')
cax = axs[1].contourf(e.results_ss_hd_snr, origin='lower', extent=[0,1,0,1])
fig.colorbar(cax, ax=axs[1], fraction=0.0485, pad=0.04)

fig.tight_layout()
```
